```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Dimensions du dataset
print("Shape:", train_data.shape)
# Visualiser les valeurs manquantes
sns.heatmap(train_data.isnull(), cbar=False, cmap="viridis")
plt.show()
```

```python
# Compter les valeurs uniques
for col in train_data.select_dtypes(include=["object"]).columns:
    print(f"{col}: {train_data[col].nunique()} catégories")
    print(train_data[col].value_counts(), "\n")
```

```python
train_data.hist(bins=20, figsize=(12, 8))
plt.tight_layout()
plt.show()
```

```python


# Exemple : scatterplot entre deux variables numériques
sns.scatterplot(x="age", y="charges", data=train_data)

# Exemple : boxplot pour voir l’impact d’une variable catégorielle
sns.boxplot(x="smoker", y="charges", data=train_data)
plt.show()

```
